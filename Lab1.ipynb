```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Creating a DataFrame using Pandas

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [25, 30, 22, 35, 28],
    'Salary': [50000, 60000, 45000, 70000, 55000]
}

df = pd.DataFrame(data)

print("DataFrame:")
df
```

```python
# NumPy Array Example

np_array = np.array([10, 20, 30, 40, 50])
print("NumPy Array:", np_array)

print("Array Mean:", np.mean(np_array))
print("Array Standard Deviation:", np.std(np_array))
```

```python
# Bar Chart for Salary Distribution

plt.figure(figsize=(8, 6))
plt.bar(df['Name'], df['Salary'])
plt.title('Salary Distribution')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.show()
```

```python
# Pie Chart for Age Distribution

plt.figure(figsize=(8, 8))
plt.pie(df['Age'], labels=df['Name'], autopct='%1.1f%%', startangle=90)
plt.title('Age Distribution')
plt.show()
```

```python
# Box Plot for Salary

plt.figure(figsize=(8, 6))
plt.boxplot(df['Salary'], vert=False)
plt.title('Salary Box Plot')
plt.xlabel('Salary')
plt.show()
```

```python
# Histogram for Age

plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, edgecolor='black')
plt.title('Age Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Subplots Example

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Line plot 1
axes[0].plot(df['Age'])
axes[0].set_title('Age Line Plot')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Age')

# Line plot 2
axes[1].plot(df['Salary'])
axes[1].set_title('Salary Line Plot')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Salary')

plt.tight_layout()
plt.show()
```
